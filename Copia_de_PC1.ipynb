Complete the tasks requested using **carbon_copy**:

PREGUNTA 1

```python
carbon_copy=carbon.loc[:,carbon.columns!='ranking']
```

iloc se usa para seleccionar por medio de los nombres, : se usó para seleccionar todas las filas, además piden seleccionar todas las columnas menos ranking, por eso se usó el !=

```python
carbon_copy
```

PREGUNTA 2

```python
Pais_min_CO2=carbon_copy.loc[carbon_copy['metric_tonnes_of_CO2'].idxmin(), 'name']
```

idxmin nos ayuda a indicar al menor valor

```python
Pais_min_CO2
```

PREGUNTA 3

```python
Pais_minCO2_por_region=carbon_copy.loc[carbon_copy.groupby('region')['metric_tonnes_of_CO2'].idxmin()]
```

groupby nos ayuda a agrupar las regios

```python
Pais_minCO2_por_region
```

```python
promedio_CO2=carbon_copy['metric_tonnes_of_CO2'].mean()
carbon_copy['CO2 - promedio']=carbon_copy['metric_tonnes_of_CO2'] - promedio_CO2
carbon_copy
```
